```python
import requests
import pandas as pd
import json
import psycopg2
```

```python
with open("token.txt") as token_do:
    token = token_do.readlines()[0]

data = {
    "token": token,
    "content": "record",
    "action": "export",
    "format": "json",
    "type": "flat",
    "csvDelimiter": "",
    "rawOrLabel": "raw",
    "rawOrLabelHeaders": "raw",
    "exportCheckboxLabel": "false",
    "exportSurveyFields": "false",
    "exportDataAccessGroups": "false",
    "returnFormat": "json",
}
r = requests.post("https://redcapdemo.vanderbilt.edu/api/", data=data)
print(f"HTTP Status: {str(r.status_code)}")
```

```python
print(f"response type: {type(r.json())}")
print(f"response length: {len(r.json())}") # number of records
print(f"response keys are the same: {[r.json()[0].keys() == r.json()[i].keys() for i in range(1,len(r.json()))]}") # All dictionaries contain the same keys
```
